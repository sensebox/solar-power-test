```python
import numpy as np
import matplotlib.pyplot as plt
import datetime

data = np.genfromtxt('power.csv', delimiter=',', dtype=None, names=['date', 'time', 'voltage'])

x = [datetime.datetime.strptime(date_str.decode('utf-8') + ' ' + time_str.decode('utf-8'), '%d.%m.%Y %H:%M') for date_str, time_str in data[['date', 'time']]]
y = data['voltage'].astype(float)

plt.figure(figsize=(10,6))
plt.plot(x, y, '-o')
plt.xticks(x, [dt.strftime('%d.%m.%Y %H:%M') for dt in x], rotation=45)
plt.xlabel('Date')
plt.ylabel('Voltage (V)')
plt.ylim(2.5, 4.2)
plt.title('Voltage vs. Date')
plt.grid(which='both')
plt.show()
```
